## Это проверка лабы по параллельному программированию. 

```python
import numpy as np
import matplotlib.pyplot as plt

RESULT_PATH = "result.txt"
MATRIX1_PATH = "matrix1.txt"
MATRIX2_PATH = "matrix2.txt"
STATS_PATH = "stats.txt"


def check():
    matrix1 = np.loadtxt(MATRIX1_PATH, int, skiprows=1)
    matrix2 = np.loadtxt(MATRIX2_PATH, int, skiprows=1)
    res_matrix = np.loadtxt(RESULT_PATH, int, skiprows=1)
    multiply = np.dot(matrix1, matrix2)
    if np.array_equal(multiply, res_matrix):
        print("Matrices are equal")
    else:
        print("Matrices are not equal")
        

def showStats(x, y, num):
    fig1 = plt.figure()
    ax1 = fig1.add_subplot(111)
    ax1.set_title(f"Время от размера ({num} потоков)")
    plt.plot(x,y, "go-")
    plt.xlabel("Matrix size")
    plt.ylabel("Milliseconds")
    for i,j in zip(x,y):
        # ax.annotate(str(j),xy=(i+10,j+100))
        ax1.text(i,j, str(j), ha='center', va='bottom')
    plt.show()


if __name__ == "__main__":
    check()
    f = open(STATS_PATH).readlines()
    f = [i.split() for i in f]
    x1 = []
    y1 = []
    x2 = []
    y2 = []
    x4 = []
    y4 = []
    x8 = []
    y8 = []
    for i in f:
        if (int(i[0]) == 1):
            x1.append(int(i[1]))
            y1.append(float(i[2]))
        if (int(i[0]) == 2):
            x2.append(int(i[1]))
            y2.append(float(i[2]))
        if (int(i[0]) == 4):
            x4.append(int(i[1]))
            y4.append(float(i[2]))
        if (int(i[0]) == 8):
            x8.append(int(i[1]))
            y8.append(float(i[2]))
    showStats(x1, y1, 1)
    showStats(x2, y2, 2)
    showStats(x4, y4, 4)
    showStats(x8, y8, 8)
```
